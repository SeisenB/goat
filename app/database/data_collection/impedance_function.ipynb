```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
from scipy.optimize import curve_fit
from scipy import optimize
from lmfit.models import StepModel, LinearModel
from sklearn.metrics import r2_score
```

```python
def gaus(x, b):
    return (np.exp(-((x)**2)/b))
def exp(x, b):
    return(np.exp(x*b))
def exp2(x, a, b):
    return(a*np.exp(b*x))
def pow(x, a):
    return(x**(-a))
```

```python
#Daily Goods
x = np.array([0.000001, 375, 750, 1500, 3750])
x2 = np.array([0.000001,187.5,562.5,1125,2625])
y = np.array([1, 0.81, 0.57, 0.21, 0.05])

#Gaussian Function
beta1 = optimize.curve_fit(gaus, xdata = x2, ydata = y, p0 = 40000)[0]
r2_1 = round(r2_score(y, gaus(x2,*beta1)),3)
beta1 = round(beta1[0])

#Negative Exponential
alpha1 = optimize.curve_fit(exp, xdata = x2, ydata = y, p0 = 0.01)[0]
r2_2 = round(r2_score(y, exp(x2,*alpha1)),3)
alpha1 = round(alpha1[0],6)

#Modified Negative Exponential
alpha2, beta2 = optimize.curve_fit(exp2, xdata = x2, ydata = y, p0 = (5.0,0.01))
r2_3 = round(r2_score(y, exp2(x2,alpha2[0],alpha2[1])),3)
alpha3 = round(alpha2[0],6)
beta2 = round(alpha2[1],6)

#plot Graphs
xmin, xmax = 0, 3000
a = plt.figure(num=1,dpi=500)
points = 1000
axes = a.add_axes([0.1,0.1,0.8,0.8])
axes.set_xlim([0,3000])
axes.set_ylim([0,1])
x2 = np.array([0,187.5,562.5,1125,2625])
y = np.array([1, 0.81, 0.57, 0.21, 0.05])
#plt.step(x1, y, where='pre')
plt.plot(x2, y, 'o--', color='grey', alpha=0.3, label='Travel Survey')
plt.title(label='Daily Goods (Food Retailing)')
axes.set_xlabel('Time [seconds]')
axes.set_ylabel('Share of trips by foot')
axes.grid(True, color ='#EEEEEE')
x = np.linspace(xmin, xmax, points)
plt.plot(x, gaus(x, beta1), label = 'Modified Gaussian\n'+'beta = '+str(beta1)+'\nR2 = '+str(r2_1))
plt.plot(x, exp(x, alpha1), label = '\nNeg. Exponential\n'+'alpha = '+str(alpha1)+'\nR2 = '+str(r2_2))
plt.plot(x, exp2(x, alpha2[0], alpha2[1]), label = '\nModif. Neg. Exp.\n'+'alpha = '+str(alpha3)+'\nbeta = '+str(beta2)+'\nR2 = '+str(r2_3))
#plt.plot(x, pow(x, alpha4[0][0]), label = 'Power')
plt.legend(loc='upper right')
plt.show()
```
